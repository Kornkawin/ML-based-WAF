# Create parameter tampering dataset

This notebook creates a custom parameter tamering dataset. We consider every parameter longer than 100 characters as parameter tampering.

```python
import numpy as np
import json
```

```python
N_SAMPLES = 1000

data = []

for i in range(N_SAMPLES//2):
    l = np.random.randint(0, 100)
    sample = {}
    sample['length'] = l
    sample['label'] = 'valid'
    data.append(sample)
```

```python
for i in range(N_SAMPLES//2):
    l = np.random.randint(100, 500)
    sample = {}
    sample['length'] = l
    sample['label'] = 'parameter-tampering'
    data.append(sample)
```

```python
data
```

```python
with open('pt_dataset.json', 'w') as f:
    json.dump(data, f)
```
